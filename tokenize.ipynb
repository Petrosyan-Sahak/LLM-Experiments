```python
import tiktoken

encoding = tiktoken.encoding_for_model("gpt-4.1-mini")
tokens = encoding.encode("let's see how is this sentence being tokenized")
```

```python
for token_id in tokens:
    token_text = encoding.decode([token_id])
    print(f"{token_id} = {token_text}")
```

```python
import os
from dotenv import load_dotenv

load_dotenv(override=True)
api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
    print(f"no api key found, exiting")
    exit(1)
else:
    print(f"found api key")
```

```python
from openai import OpenAI

openai = OpenAI(base_url="http://localhost:11434/v1", api_key=api_key)
```

```python
messages = [
    {"role": "system", "content": "This is the first message"},
    {"role": "user", "content": "how old are you?"}
]
```

```python
response = openai.chat.completions.create(model="gemma3:270m", messages=messages)
response.choices[0].message.content
```
